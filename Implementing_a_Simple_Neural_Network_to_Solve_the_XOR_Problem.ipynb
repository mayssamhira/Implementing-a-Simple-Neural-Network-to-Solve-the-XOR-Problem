```python

import numpy as np
from keras.models import Sequential
from keras.layers import Dense
from tensorflow.keras.utils import to_categorical
```

```python

# Define the model
model = Sequential([
    Dense(4, input_dim=2, activation='relu'),  # Hidden layer
    Dense(1, activation='sigmoid')  # Output layer
])

```

```python
# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

```

```python
# Dataset for XOR problem
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([[0], [1], [1], [0]])
```

```python

# Train the model
model.fit(X, y, epochs=100)
```

```python
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder

data = pd.read_csv('Data.csv')
features = data[['income','age','loan']].values
labels = data['default'].values.reshape(-1,1)
encoder = OneHotEncoder()
encoded_labels = encoder.fit_transform(labels).toarray()
X_train, X_test, y_train, y_test = train_test_split(features, encoded_labels, test_size=0.2, random_state=42)

model = Sequential([
    Dense(10, input_shape=(3,), activation='relu'),
    Dense(2, activation='softmax')
])
optimizer = Adam(learning_rate=0.0005)
model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=1000, batch_size=10)

loss, accuracy = model.evaluate(X_test, y_test)
print(f'Test Loss: {loss}, Test Accuracy: {accuracy}')
```
